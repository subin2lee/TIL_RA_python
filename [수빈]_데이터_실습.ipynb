### merge

두 개의 데이터를 합쳐서 보고 싶을 때 사용함  
book_tags에 tags을 merge함  
어떤 columne 기준으로 할 것인지는 <b>on=~~에 작성한다





```python
book_tags = book_tags.merge(tags, on = 'tag_id')
```

```python
book_tags.columns
```

### 데이터를 특정을 하고 싶을 때  
부분적으로 사용하여 merge할 경우  
대괄호 안에 컬럼의 리스트들을 넣어주면 됨

```python
book_tags = book_tags.merge(books[['goodreads_book_id', 'title']], on = 'goodreads_book_id')
```

### loc
어떤 위치에 있는 얘들을 고쳐줘야하는지에 대해서 located하는 것  
[어떤 row에 해당하는지(조건문), 해당 열의 columns명]

```python
# book_tags의 count 칼럼 중 0보다 작은 얘들의 값을 count 값에 0을 넣어준다는 의미
book_tags.loc[book_tags['count'] < 0, 'count'] = 0
```

```python
book_tags.describe()
```

# Visualization
데이터 시각화는 처음 본 사람들이 직관적으로 이를 이해하는데 좋음  
seaborn이 시각적으로 이쁘대요 ^_^*    

-- 차이점은 구글링을 하기 --  
그래프의 명칭을 이용해서 검색해보면 더 좋음

```python
import matplotlib.pyplot as plt
import seaborn as sns

top_rated = books.sort_values('average_rating', ascending=False)
tf_top_rated = top_rated[:25] # 인덱싱은 중요함 (구글링하래,)

# Figure 크기 설정 (우리가 알아서 설정가능함)
plt.figure(figsize=(15, 7))

# Seaborn으로 수평 막대 그래프 그리기
sns.barplot(x="average_rating", y="original_title", data=tf_top_rated, palette="viridis")

# 그래프 제목 설정
plt.title('Top Rated Books and Their Ratings', fontsize = 16)

# x축, y축 라벨 설정
plt.xlabel('Average Rating', fontsize = 14)
plt.ylabel('Book Title', fontsize = 14)

# x축 값 범위 설정 (선택사항 - 컴퓨터가 알아서 해줌, 하지만 사용자가 원하는 범위가 있으면 알아서...)
plt.xlim(0, 5)

# y축 라벨 회전 (필요한 경우)
plt.yticks(rotation=0, fontsize=12)

# 그래프 표시
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

top_popular = books.sort_values('ratings_count', ascending=False)
tf_top_popular = top_popular[:20] # 인덱싱은 중요함 (구글링하래,)

# Figure 크기 설정 (우리가 알아서 설정가능함)
plt.figure(figsize=(15, 7))

# Seaborn으로 수평 막대 그래프 그리기
sns.barplot(x="ratings_count", y="original_title", data=tf_top_popular, palette="viridis")

# 그래프 제목 설정
plt.title('Top Popular Books and Their Counts', fontsize = 16)

# x축, y축 라벨 설정
plt.xlabel('Rating Count', fontsize = 14)
plt.ylabel('Book Title', fontsize = 14)

# x축 값 범위 설정 (선택사항)
plt.xlim(0, 5)

# y축 라벨 회전 (필요한 경우)
plt.yticks(rotation=0, fontsize=12)

# 그래프 표시
plt.show()
```

```python
from wordcloud import WordCloud, STOPWORDS
```

```python
authoor_string = " ".join(books['authors'])
title_string = " ".join([str(book_name) for book_name in books['original_title']])
```

```python
def wordcloud(string) :
    stopwords = set(STOPWORDS) # 사용하지 않는 단어를 걸러줌
    wc = WordCloud(width=800, height=500, mask=None, random_state=21, # 동일값을 가지게 하기 위해서 random_state를 해줌
                 max_font_size=110, stopwords=stopwords).generate(string) # 관사와 같은 것도 걸러줌
    fig = plt.figure(figsize=(16, 8))
    plt.axis('off')
    plt.imshow(wc)
```

```python
wordcloud(authoor_string)
```
